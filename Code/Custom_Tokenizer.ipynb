<h2 class="alert alert-block alert-success">Reads the Verdict Novel.</h2>

<div class="alert alert-block alert-success">
"The Verdict" novel is read which contains 20479 words.

```python
file_path = '../Data/the-verdict.txt'
with open(file_path, 'r', encoding='utf-8') as f:
    raw_text = f.read()
print(len(raw_text))
print(raw_text[:100])
```

<div class="alert alert-block alert-success">
Extract out the unique words from the novel which is 1130.
</div>


```python
import re

expression = r'([,.:;?_!"()\']|--|\s)'
preprocessed = re.split(expression, raw_text)
print(preprocessed[:30])
preprocessed = [word for word in preprocessed if word.split()]
print(preprocessed[:30])
```

```python
all_words = sorted(set(preprocessed))
print(len(all_words))
```

<h2 class="alert alert-block alert-success">Encoding</h2>
<div class="alert alert-block alert-success">
Map sorted token or words to Integer. This process is called Encoding
</div>

```python
from itertools import islice

vocab = {token: index for index, token in enumerate(all_words)}
print("Encoding: Word -> Integer mapping\n\n", vocab)

reverse = {value:key for key, value in vocab.items()}
print("Integer -> Word mapping\n\n", reverse)
```

<h2 class="alert alert-block alert-success">Tokenizer V1 class</h2>

```python
import re
from typing import List
from collections import defaultdict

class TokenizerV1:
    # Tokenizer init with words which are part of Vocabulary.
    # It splits them, put them in set to ensure a word is considered only once.
    # Create mapping of word -> Integer
    def __init__(self, vocab):
        self.expression = r'([,.:;?_!"()\']|--|\s)'
        words = re.split(self.expression, vocab)
        all_unique_words = [word.strip() for word in sorted(set(words
        )) if word.strip()]
        self.str_to_int = {word.strip():index for index, word in enumerate(all_unique_words)}
        self.int_to_str = {index:word for word, index in self.str_to_int.items()}
    
    def encode(self, text):
        words = re.split(self.expression, text)
        all_unique_words = [word.strip() for word in words if word.strip()]
        ids = [ self.str_to_int[word] for word in all_unique_words]
        print(ids[:10])
        return ids

    def decode(self, ids):
        text = " ".join(self.int_to_str[id] for id in ids)
        text = re.sub(self.expression, r'\1', text)
        print(text)
    

with open(file_path, 'r', encoding='utf-8') as f:
    raw_text = f.read()
print(len(raw_text))
print(raw_text[:100])

tokenizerV1 = TokenizerV1(raw_text)
ids = tokenizerV1.encode("Jack Gisburn rather?")
tokenizerV1.decode(ids)
```
